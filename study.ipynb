```python
from tcnqs import toy_ham
import numpy as np
import matplotlib.pyplot as plt

# one can play with the parameters of the Hamiltonian, see toy_ham.py for details
ham = toy_ham.init_hamiltonian(n=10, degen_n=0, delta=0.5, m_off=0.4, nnz_ratio=0.5)
print(ham)

# diagonalize the Hamiltonian and find the ground state
eigvals, eigvecs = np.linalg.eigh(ham)
# sort the eigenvalues and eigenvectors, according to descending order of eigenvalues
idx = eigvals.argsort()
eigvals = eigvals[idx]
print(eigvals)
eigvecs = eigvecs[idx,:]
gs = eigvecs[0, :]

# make sure the first c_0 is positive
gs = np.sign(gs[0]) * gs

# plot the ground state vector 
plt.plot(gs, 'o-')

plt.xlabel('Configuration Index')
plt.ylabel('C$_i$ Amplitude')
```
